### Correlation Map
Finding the correlation map using the seaborn library

```python
import seaborn as sb #Statistical data visualization

f,ax = plt.subplots(figsize=(18,18))
sb.heatmap(data.corr(), annot=True, linewidths=.5, fmt='.1f', ax=ax)
```
